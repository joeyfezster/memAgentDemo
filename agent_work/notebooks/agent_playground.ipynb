# Multi-Agent Playground
Use this notebook to tinker with the routing, specialized, and generalist agents built with Letta.

```python
from pathlib import Path
import sys
import asyncio
REPO_ROOT = Path('..').resolve().parent
BACKEND_ROOT = REPO_ROOT / 'backend'
if str(BACKEND_ROOT) not in sys.path:
    sys.path.append(str(BACKEND_ROOT))
```

```python
from app.agents import AgentRequestContext, get_agent_orchestrator

async def interact(prompt: str):
    orchestrator = get_agent_orchestrator()
    await orchestrator.bootstrap()
    context = AgentRequestContext(
        user_id='notebook-user',
        persona_handle='dev',
        display_name='Notebook Dev',
    )
    reply = await orchestrator.route(context, prompt)
    return reply

async def run_demo():
    sample = 'Compare current retention against the prior release and call out drivers.'
    return await interact(sample)

demo_reply = asyncio.run(run_demo())
demo_reply
```

```python
from app.agents.evals import run_eval, summarize_eval

async def run_default_eval():
    results = await run_eval()
    return summarize_eval(results)

evaluation_report = asyncio.run(run_default_eval())
evaluation_report
```
